```python
from Bio import SeqIO
#import numpy as np

def levenshtein_distance(a, b):
    """Levenshtein distance between two strings"""
    if not a: return len(b)
    if not b: return len(a)
    prev_row = range(len(b) + 1)
    for i, ch1 in enumerate(a):
        current_row = [i + 1]
        for j, ch2 in enumerate(b):
            insertions = prev_row[j + 1] + 1
            deletions = current_row[j] + 1
            substitutions = prev_row[j] + (ch1 != ch2)
            current_row.append(min(insertions, deletions, substitutions))
        prev_row = current_row

    return prev_row[-1]

original_ids = {record.description for record in SeqIO.parse("/home/majnusova/all/projects/bilabrum/data/MCPs_PLV_EVEs_confirmed.faa", "fasta")}
modified_ids = {record.description for record in SeqIO.parse("/home/majnusova/all/projects/bilabrum/data/bellas_renamed.fasta", "fasta")}

# Pro každé modifikované ID najdeme originální ID s nejnižší Levenshteinovou vzdáleností
closest_matches = {}
for mod_id in modified_ids:
    closest_original = min(original_ids, key=lambda orig_id: levenshtein_distance(mod_id, orig_id))
    closest_matches[mod_id] = closest_original
```

```python
from Bio import SeqIO
#import numpy as np

def levenshtein_distance(a, b):
    """Levenshtein distance between two strings"""
    if not a: return len(b)
    if not b: return len(a)
    prev_row = range(len(b) + 1)
    for i, ch1 in enumerate(a):
        current_row = [i + 1]
        for j, ch2 in enumerate(b):
            insertions = prev_row[j + 1] + 1
            deletions = current_row[j] + 1
            substitutions = prev_row[j] + (ch1 != ch2)
            current_row.append(min(insertions, deletions, substitutions))
        prev_row = current_row

    return prev_row[-1]

original_ids = {record.description for record in SeqIO.parse("/home/majnusova/all/projects/bilabrum/data/GVOGm0003.copy1_MCP.faa", "fasta")}
modified_ids = {record.description for record in SeqIO.parse("/home/majnusova/Desktop/ncldv_renamed.fasta", "fasta")}

# Pro každé modifikované ID najdeme originální ID s nejnižší Levenshteinovou vzdáleností
closest_matches = {}
for mod_id in modified_ids:
    closest_original = min(original_ids, key=lambda orig_id: levenshtein_distance(mod_id, orig_id))
    closest_matches[mod_id] = closest_original
```

```python
from Bio import SeqIO
gvogs = {}
with open("/home/majnusova/all/projects/bilabrum/data/GVOGm0760.copy1_A32.faa", "r") as infile:
    for record in SeqIO.parse(infile, "fasta"):
        header = record.description
        modified_header = header.replace(" ", "_")
        gvogs[header] = modified_header
gvogs

gvogs_modified = {}
with open("/home/majnusova/Desktop/ncldv_renamed.fasta", "r") as inf:
    for record in SeqIO.parse(inf, "fasta"):
        id = record.description
        seq = record.seq
        gvogs_modified[id] = seq
gvogs_modified 

id_pairs = {}
for origo, modif in gvogs.items():
    for id in gvogs_modified.keys():
        if id in modif:
            id_pairs[id] = origo
id_pairs
with open("/home/majnusova/all/projects/bilabrum/data/ncldvs_ids_pairs.txt", "w") as outf:
    for i, j in id_pairs.items():
        outf.write(f"{i}\t{j}\n")
```

```python
from Bio import SeqIO
import pandas as pd

# Načtení hlaviček z FASTA souboru
headers = []
with open("/home/majnusova/Desktop/virocluster.fasta", "r") as infile:
    for record in SeqIO.parse(infile, "fasta"):
        headers.append(record.description)

# Načtení tabulky z Excel souboru
tbl = pd.read_excel("/home/majnusova/Desktop/mcp.xlsx")

# Filtrování DataFrame tak, aby zůstaly pouze řádky s ID v sloupci 'Modified', které odpovídají 'headers'
found_mask = tbl['Modified'].isin(headers)
found_rows = tbl.loc[found_mask, ['Modified', 'data_ids', 'data_source', 'data_type', 'prot_seq']]

# Uložení nalezených řádků do souboru CSV
output_filepath = "/home/majnusova/Desktop/found_ids_and_values.csv"
found_rows.to_csv(output_filepath, index=False, sep = "\t")

print(f"Nalezená ID a hodnoty byly uloženy do souboru: {output_filepath}")
```
